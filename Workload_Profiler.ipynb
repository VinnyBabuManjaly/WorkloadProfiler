# Workload Profiler

## Project Overview

This project presents a data-driven framework for automatically discovering and characterizing distinct workload types in large-scale compute clusters. Using the Google 2019 Cluster Sample (derived from Google’s production cluster traces), the system analyzes job and task-level resource requests, observed CPU and memory utilization patterns, runtime characteristics, and scheduling metadata to identify meaningful behavioral segments.

By transforming raw time-series usage data into structured workload profiles, the project applies dimensionality reduction and unsupervised clustering techniques to uncover natural workload categories. These categories may include short, bursty batch jobs, long-running latency-sensitive services, memory-intensive analytics tasks, and other operationally distinct patterns.

The primary objective is to provide actionable insights for infrastructure and network operations teams. By understanding how workloads differ in resource intensity, variability, and efficiency, stakeholders can design differentiated scheduling policies, optimize capacity planning, and implement predictive autoscaling strategies tailored to each workload type.

Ultimately, this approach enhances resource utilization, reduces overprovisioning, mitigates performance bottlenecks, and supports more intelligent, behavior-aware cluster management.

```python
# -----------------------------
# 3.2 Clean Data
# -----------------------------

df_clean = df_selected.copy()

# --- 1. Handle missing values ---

# Fill missing vertical_scaling with 0 (no scaling)
if "vertical_scaling" in df_clean.columns:
    df_clean["vertical_scaling"] = df_clean["vertical_scaling"].fillna(0)

# Fill missing scheduler with mode (most frequent value)
if "scheduler" in df_clean.columns:
    df_clean["scheduler"] = df_clean["scheduler"].fillna(
        df_clean["scheduler"].mode()[0]
    )

# Dropping missing rows with missing'resource_request' field ---
df_clean = df_clean.dropna(subset=["resource_request"])

# --- 2. Parse JSON resource usage columns into numeric CPU/memory ---

def extract_cpu_memory(value):
    try:
        data = json.loads(str(value).replace("'", '"'))
        return pd.Series([data.get("cpu", 0), data.get("memory", 0)])
    except Exception:
        return pd.Series([0, 0])

# Average usage
df_clean[["avg_cpu", "avg_memory"]] = df_clean["average_usage"].apply(extract_cpu_memory)

# Maximum usage
df_clean[["max_cpu", "max_memory"]] = df_clean["maximum_usage"].apply(extract_cpu_memory)

# Parse resource_request into numeric columns
df_clean[['req_cpu', 'req_memory']] = df_clean["resource_request"].apply(extract_cpu_memory)

# Drop raw JSON and distribution columns (if present)
df_clean = df_clean.drop(
    columns=[
        "average_usage",
        "maximum_usage",
        "cpu_usage_distribution",
        "tail_cpu_usage_distribution",
        "resource_request"
    ],
    errors="ignore",
)

# --- 3. Compute runtime_seconds for later modeling/cleaning use ---

df_clean["start_time"] = pd.to_datetime(df_clean["start_time"])
df_clean["end_time"] = pd.to_datetime(df_clean["end_time"])

df_clean["runtime_seconds"] = (
    df_clean["end_time"] - df_clean["start_time"]
).dt.total_seconds()

# We already filtered end_time > start_time in Select Data, so runtime_seconds > 0 by design.

# --- 4. Handle outliers in key numeric fields ---

numeric_cols = ["avg_cpu", "max_cpu", "req_cpu", "assigned_memory", "runtime_seconds"]

for col in numeric_cols:
    if col in df_clean.columns:
        lower = df_clean[col].quantile(0.01)
        upper = df_clean[col].quantile(0.99)
        df_clean[col] = df_clean[col].clip(lower=lower, upper=upper)

print("Clean dataset shape:", df_clean.shape)

# Optional: check remaining missing values
df_clean.isnull().sum()
```

### 3.3 Construct Data


In this step, we create derived attributes from the cleaned dataset to better capture workload characteristics for clustering and profiling. These new features emphasize resource efficiency, overprovisioning behavior, and usage patterns that directly relate to workload types.

The construction focuses on ratios and efficiency metrics rather than raw values, as these normalized measures are more stable across different scales and better reveal behavioral patterns in distance-based clustering.

#### Derived Attributes Created

Resource Efficiency & Overprovisioning Features:

- `memory_overprovisioning_ratio` = `assigned_memory` / `resource_request`  
  Measures how much more memory was assigned than requested (values > 1 indicate overprovisioning).

- `avg_cpu_utilization` = `avg_cpu` / `assigned_memory`  
  CPU utilization relative to assigned resources (captures underutilization).

- `peak_cpu_utilization` = `max_cpu` / `assigned_memory`  
  Peak CPU demand relative to assigned capacity.

- `memory_utilization_avg` = `avg_memory` / `assigned_memory`  
  Average memory utilization efficiency.

- `memory_utilization_peak` = `max_memory` / `assigned_memory`  
  Peak memory utilization efficiency.

Workload Intensity & Variability:

- `cpu_peak_to_avg_ratio` = `max_cpu` / `avg_cpu`  
  Indicates bursty vs steady CPU workloads (higher values = more bursty).

- `runtime_efficiency` = `runtime_seconds` / `assigned_memory`  
  Runtime normalized by resource allocation (long-running low-resource jobs vs short high-resource jobs).

Page Cache Dependency:

- `page_cache_ratio` = `page_cache_memory` / `assigned_memory`  
  Proportion of assigned memory used for caching (IO-intensive workloads tend to have higher values).

These derived attributes transform the raw allocation/usage data into behavioral signals that are more suitable for discovering workload types through clustering.

### 4.3 Build Model

#### Generic Functions

**Dunn Index Function** 

This function computes the Dunn Index, a metric to evaluate clustering quality.

It:

* Ignores noise points (`-1` labels).
* Checks early exit: returns NaN if fewer than 2 clusters.
* For large datasets (>10k samples), it skips the expensive calculation and approximates using the silhouette score.
* Computes:

  * Intra-cluster distances: maximum distance within each cluster (compactness).
  * Inter-cluster distances: minimum distance between clusters (separation).
* Returns the Dunn Index: min inter-cluster distance ÷ max intra-cluster distance.

Higher values : well-separated, tight clusters.


```python
# Global results storage
all_results = []

def dunn_index(X, labels):
    """Dunn Index: min(inter-cluster dist) / max(intra-cluster dist)"""
    unique_labels = [l for l in np.unique(labels) if l != -1]
    if len(unique_labels) < 2:
        return np.nan
    
    n_samples = X.shape[0]
    
    # Skip full computation for large datasets (>10k samples)
    if n_samples > 10000:
        print("Dataset too large for exact Dunn Index - using approximation")
        # Simple heuristic: use silhouette as proxy (both measure separation)
        return silhouette_score(X, labels) * 0.8  # Approximate scaling
    
    D = pairwise_distances(X)
    intra_dists = []
    for c in unique_labels:
        idx = np.where(labels == c)[0]
        if len(idx) > 1:
            intra = D[np.ix_(idx, idx)]
            intra_dists.append(np.max(intra))
    
    inter_dists = []
    for i, ci in enumerate(unique_labels):
        for cj in unique_labels[i+1:]:
            idx_i = np.where(labels == ci)[0]
            idx_j = np.where(labels == cj)[0]
            inter = D[np.ix_(idx_i, idx_j)]
            inter_dists.append(np.min(inter))
    
    return np.min(inter_dists) / np.max(intra_dists) if inter_dists and intra_dists else np.nan
```

**Cluster size distribution statistics**

This function summarizes the cluster size distribution.

* Removes noise points (label `-1`).
* If no valid clusters remain, it returns zeros/NaN.
* Counts how many clusters exist.
* Calculates:
  - Total number of clusters
  - Largest cluster size (as % of data)
  - Smallest cluster size (as % of data)

Overall, it helps check whether clusters are balanced or if one cluster dominates the dataset.

```python

def cluster_size_stats(labels):
    """Cluster size distribution statistics"""
    labels_arr = np.array(labels)
    mask = labels_arr != -1
    labels_no_noise = labels_arr[mask]
    
    if len(labels_no_noise) == 0:
        return {"n_clusters": 0, "max_cluster_pct": np.nan, "min_cluster_pct": np.nan}
    
    unique, counts = np.unique(labels_no_noise, return_counts=True)
    total = len(labels_no_noise)
    return {
        "n_clusters": int(len(unique)),
        "max_cluster_pct": float(np.max(counts) / total),
        "min_cluster_pct": float(np.min(counts) / total)
    }
```

```python
#----------------------------------------------------
# Generic Visualization Function
# ----------------------------------------------------
def visualize_clustering_results(X, labels, result):
    """
    Enhanced 2x2 visualization for clustering results with clear labels and details
    """
    model_name = result['model_name']
    metrics = result['metrics']
    size_stats = result['size_stats']
    n_clusters = result['n_clusters']
    noise_pct = result['noise_points'] / len(labels) * 100
    
    # Create figure with better spacing
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    fig.suptitle(f'{model_name}\n{n_clusters} Clusters | Noise: {noise_pct:.1f}%', 
                 fontsize=18, fontweight='bold', y=0.98)
    
    # 1. Cluster Size Distribution
    ax = axes[0, 0]
    valid_labels = labels[labels != -1]
    if len(valid_labels) > 0:
        unique, counts = np.unique(valid_labels, return_counts=True)
        total_valid = len(valid_labels)
        bars = ax.bar(unique, counts, color='steelblue', alpha=0.8, edgecolor='navy')
        ax.set_title('Cluster Size Distribution\n(Noise Points Excluded)', fontsize=12, fontweight='bold')
        ax.set_xlabel('Cluster ID', fontweight='bold')
        ax.set_ylabel('Number of Workloads', fontweight='bold')
        
        # Add value labels on bars
        for i, (bar, count) in enumerate(zip(bars, counts)):
            pct = count / total_valid * 100
            ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + max(counts)*0.01, 
                   f'{count:,}\n({pct:.1f}%)', ha='center', va='bottom', fontweight='bold')
        
        # Add summary text
        ax.text(0.02, 0.98, f'Total: {total_valid:,} workloads\n{n_clusters} clusters', 
                transform=ax.transAxes, fontsize=10, verticalalignment='top',
                bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))
    else:
        ax.text(0.5, 0.5, 'No Valid Clusters\nAll Points Classified as Noise', 
                ha='center', va='center', fontsize=14, fontweight='bold')
        ax.set_title('Cluster Size Distribution', fontweight='bold')
    
    ax.grid(True, alpha=0.3, axis='y')
    
    # 2. Key Clustering Metrics
    ax = axes[0, 1]
    metric_names = list(metrics.keys())
    metric_values = list(metrics.values())
    
    if metric_names:
        colors = plt.cm.Set3(np.linspace(0, 1, len(metric_names)))
        bars = ax.bar(metric_names, metric_values, color=colors, alpha=0.8, edgecolor='black')
        ax.set_title('Clustering Quality Metrics\n(Higher=better except Davies-Bouldin)', 
                    fontsize=12, fontweight='bold')
        ax.set_ylabel('Metric Score', fontweight='bold')
        ax.tick_params(axis='x', rotation=45)
        
        # Add value labels
        max_val = max(metric_values) if metric_values else 1
        for bar, value in zip(bars, metric_values):
            height = bar.get_height()
            ax.text(bar.get_x() + bar.get_width()/2., height + max_val*0.02,
                   f'{value:.3f}', ha='center', va='bottom', fontweight='bold')
        
        # Add metric interpretation guide
        ax.text(0.02, 0.02, '↑↑↑ Better    ↓↓↓ Worse', transform=ax.transAxes, 
                fontsize=9, style='italic', color='darkred')
    else:
        ax.text(0.5, 0.5, 'Metrics Not Computed\n(Single Cluster or All Noise)', 
                ha='center', va='center', fontsize=14, fontweight='bold')
        ax.set_title('Clustering Quality Metrics', fontweight='bold')
    
    ax.grid(True, alpha=0.3, axis='y')
    
    # 3. PCA 2D Visualization
    ax = axes[1, 0]
    pca = PCA(n_components=2, random_state=42)
    X_pca = pca.fit_transform(X)
    
    unique_labels = np.unique(labels)
    colors = plt.cm.tab20(np.linspace(0, 1, len(unique_labels)))
    
    scatter = ax.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='tab20', 
                        s=30, alpha=0.7, edgecolors='w', linewidth=0.5)
    
    ax.set_title(f'PCA 2D Projection of Workload Features\nPC1+PC2: {pca.explained_variance_ratio_.sum():.1%} Variance', 
                fontsize=12, fontweight='bold')
    ax.set_xlabel(f'Principal Component 1\n({pca.explained_variance_ratio_[0]:.1%} variance)', fontweight='bold')
    ax.set_ylabel(f'Principal Component 2\n({pca.explained_variance_ratio_[1]:.1%} variance)', fontweight='bold')
    
    # ---- Print PCA Results ----
    print("\n🔎 PCA Summary")
    print("-" * 50)
    print(f"Explained Variance Ratio (PC1): {pca.explained_variance_ratio_[0]:.4f}")
    print(f"Explained Variance Ratio (PC2): {pca.explained_variance_ratio_[1]:.4f}")
    print(f"Total Variance Explained (PC1+PC2): {pca.explained_variance_ratio_.sum():.4f}")
    
    print("\nPrincipal Component Loadings (Top Drivers):")
    for i, component in enumerate(pca.components_):
        sorted_idx = np.argsort(np.abs(component))[::-1][:5]  # top 5 strongest features
        print(f"\nPC{i+1} Top Features:")
        for idx in sorted_idx:
            print(f"  Feature {idx}: {component[idx]:.4f}")

    # Custom colorbar with cluster count
    cbar = plt.colorbar(scatter, ax=ax, shrink=0.8)
    cbar.set_label('Cluster Label (-1 = Noise)', rotation=270, labelpad=20, fontweight='bold')
    
    # Add sample count annotation
    ax.text(0.02, 0.02, f'N={len(X):,}', transform=ax.transAxes, fontsize=12,
            bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
    
    # 4. Silhouette Score Analysis
    ax = axes[1, 1]
    if 'silhouette' in metrics and not np.isnan(metrics['silhouette']):
        from sklearn.metrics import silhouette_samples
        mask_valid = labels != -1
        if mask_valid.sum() > 0:
            sil_samples = silhouette_samples(X[mask_valid], labels[mask_valid])
            
            sns.histplot(sil_samples, bins=min(30, len(sil_samples)//10), kde=True, 
                        ax=ax, color='teal', alpha=0.7, edgecolor='black')
            
            # Mean silhouette line
            mean_sil = metrics['silhouette']
            ax.axvline(mean_sil, color='red', linestyle='--', linewidth=3, 
                      label=f'Mean Silhouette\n{mean_sil:.3f}')
            ax.axvline(0, color='gray', linestyle='-', alpha=0.5)
            
            ax.set_title('Silhouette Score Distribution\n(>0 = good cluster membership)', 
                        fontsize=12, fontweight='bold')
            ax.set_xlabel('Silhouette Score (-1 to +1)', fontweight='bold')
            ax.set_ylabel('Number of Workloads', fontweight='bold')
            ax.legend(fontsize=10, loc='upper left')
            ax.grid(True, alpha=0.3)
            
            # Add quality indicator
            quality = "Excellent" if mean_sil > 0.6 else "Good" if mean_sil > 0.4 else "Fair"
            color = 'green' if mean_sil > 0.4 else 'orange' if mean_sil > 0.2 else 'red'
            ax.text(0.98, 0.98, f'Quality: {quality}', transform=ax.transAxes,
                   fontsize=12, fontweight='bold', color=color,
                   bbox=dict(boxstyle='round', facecolor='white', alpha=0.9),
                   ha='right', va='top')
        else:
            ax.text(0.5, 0.5, 'No Valid Points\nfor Silhouette', ha='center', va='center')
    else:
        ax.text(0.5, 0.5, 'Silhouette Analysis\nNot Available\n(Single Cluster)', 
                ha='center', va='center', fontsize=14, fontweight='bold')
        ax.set_title('Silhouette Score Analysis', fontweight='bold')
    
    plt.tight_layout()
    
    # Save with timestamp and model info
    safe_name = model_name.lower().replace(' ', '_').replace('(', '').replace(')', '').replace(',', '')
    plt.savefig(f"plots/clustering_eval_{safe_name}.png", dpi=300, bbox_inches='tight', 
                facecolor='white')
    plt.show()

    print(f"📊 Plot saved: plots/clustering_eval_{safe_name}.png")
```

**Generic Clustering Evaluation Function**

This function is a complete clustering evaluation wrapper.

It:

* Fits the clustering model (or uses precomputed labels for baselines).
* Measures fit time.
* Counts clusters and noise points.
* Computes quality metrics (Silhouette, Davies–Bouldin, Calinski–Harabasz, Dunn) when valid.
* Calculates cluster size statistics (largest and smallest cluster %).
* Stores all results in a structured dictionary for comparison.
* Optionally generates the visualization dashboard.

In short, it standardizes how every clustering model in the project is evaluated, compared, and reported.


```python

# ----------------------------------------------------
# Generic Clustering Evaluation Function
# ----------------------------------------------------
def evaluate_clustering_model(model_or_labels, X, model_name, labels=None, visualize=True):
    """
    Comprehensive clustering model evaluation - HDBSCAN, DBSCAN, KMeans, Baselines
    """
    global all_results
    
    print(f"\n{'='*80}")
    print(f"Evaluating Clustering Model: {model_name}")
    print('='*80)
    
    X_arr = np.asarray(X)
    
    if labels is not None:
        # Baseline case - use provided labels directly
        labels = np.asarray(labels)
        fit_time = 0.001  # Negligible computation
        print("Using pre-computed baseline labels")
    else:
        # Scikit-learn model case
        start_time = time.time()
        try:
            labels = model_or_labels.fit_predict(X_arr)
        except AttributeError:
            model_or_labels.fit(X_arr)
            labels = model_or_labels.labels_
        fit_time = time.time() - start_time
    
    labels = np.asarray(labels)
    unique_labels = np.unique(labels)
    n_clusters = np.sum(unique_labels != -1)
    
    print(f"Fit Time: {fit_time:.3f} seconds")
    print(f"Clusters found: {n_clusters}")
    print(f"Noise points: {np.sum(labels == -1)}")
    print(f"Label distribution: {dict(zip(*np.unique(labels, return_counts=True)))}")
    
    # Compute metrics (only if >1 cluster)
    metrics = {}
    if n_clusters > 1:
        mask_valid = labels != -1
        X_valid = X_arr[mask_valid]
        labels_valid = labels[mask_valid]
        
        if len(np.unique(labels_valid)) > 1:
            metrics = {
                'silhouette': silhouette_score(X_valid, labels_valid),
                'davies_bouldin': davies_bouldin_score(X_valid, labels_valid),
                'calinski_harabasz': calinski_harabasz_score(X_valid, labels_valid),
                'dunn_index': dunn_index(X_valid, labels_valid)
            }
    
    # Model-specific metrics
    model_metrics = {}
    if hasattr(model_or_labels, 'inertia_'):
        model_metrics['inertia'] = model_or_labels.inertia_
    if hasattr(model_or_labels, 'eps_'):
        model_metrics['eps'] = getattr(model_or_labels, 'eps_')
    
    # Cluster size stats
    size_stats = cluster_size_stats(labels)
    
    # Print results
    print(f"\nClustering Quality Metrics:")
    for metric, value in metrics.items():
        print(f"  {metric.replace('_', ' ').title()}: {value:.4f}")
    
    print(f"\nCluster Stats:")
    print(f"  Clusters: {size_stats['n_clusters']}")
    print(f"  Largest cluster: {size_stats['max_cluster_pct']:.1%}")
    print(f"  Smallest cluster: {size_stats['min_cluster_pct']:.1%}")
    
    # Store results
    result = {
        'model_name': model_name,
        'labels': labels,
        'fit_time': fit_time,
        'n_clusters': n_clusters,
        'metrics': metrics,
        'model_metrics': model_metrics,
        'size_stats': size_stats,
        'noise_points': int(np.sum(labels == -1))
    }
    all_results.append(result)
    
    if visualize:
        visualize_clustering_results(X_arr, labels, result)
    
    print(f"\n✅ {model_name} evaluation complete")
    return result
```

```python
# ----------------------------------------------------
# Analysis (Summary Table)
# ----------------------------------------------------
def print_summary_table():
    """Print comparison table of all results"""
    if not all_results:
        print("No results to summarize")
        return
    
    summary_data = []
    for result in all_results:
        row = {
            'Model': result['model_name'],
            'Clusters': result['n_clusters'],
            'Fit Time': f"{result['fit_time']:.2f}s",
            'Silhouette': f"{result['metrics'].get('silhouette', np.nan):.3f}",
            'DB Index': f"{result['metrics'].get('davies_bouldin', np.nan):.3f}",
            'CH Score': f"{result['metrics'].get('calinski_harabasz', np.nan):.0f}",
            'Noise %': f"{result['noise_points']/len(result['labels']):.1%}",
            'Max Cluster': f"{result['size_stats']['max_cluster_pct']:.1%}"
        }
        summary_data.append(row)
    
    summary_df = pd.DataFrame(summary_data)
    print("\n" + "="*120)
    print("CLUSTERING MODEL COMPARISON SUMMARY")
    print("="*120)
    print(summary_df.to_string(index=False))
```
